3 Outlier Detection:

```python
#detecting GDP and score outliers
# Outliers for 'score'
Q1_s = np.percentile(south_asian_data['score'], 25)
Q3_s = np.percentile(south_asian_data['score'], 75)
IQR_s = Q3_s - Q1_s
lb_s = Q1_s - 1.5 * IQR_s
ub_s = Q3_s + 1.5 * IQR_s
out_s = south_asian_data[(south_asian_data['score'] < lb_s) | (south_asian_data['score'] > ub_s)]

# Outliers for 'Log GDP per capita'
Q1_g = np.percentile(south_asian_data['Log GDP per capita'], 25)
Q3_g = np.percentile(south_asian_data['Log GDP per capita'], 75)
IQR_g = Q3_g - Q1_g
lb_g = Q1_g - 1.5 * IQR_g #lower bound
ub_g = Q3_g + 1.5 * IQR_g #upper bound
out_g = south_asian_data[(south_asian_data['Log GDP per capita'] < lb_g) | (south_asian_data['Log GDP per capita'] > ub_g)]

# Concise output
print("Score outliers:", out_s['Country name'].tolist()) 
print("GDP per capita outliers:", out_g['Country name'].tolist())
```

```python
#plotting
plt.scatter(south_asian_data['Log GDP per capita'], south_asian_data['score'], label='Normal', color='green')
plt.scatter(out_g['Log GDP per capita'], out_s['score'], label='Outliers', color='red')

plt.xlim(0.5, 1.5)  
plt.ylim(0, 5.5)

plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.title('South Asia:GDP per Capita vs Score')
plt.legend()
plt.show()
```

6. Outlier Detection

```python
# Function to identify outliers using IQR method
def identify_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Identify outliers for South Asia
south_asia_score_outliers = identify_outliers(south_asian_data, 'score')
south_asia_gdp_outliers = identify_outliers(south_asian_data, 'Log GDP per capita')

# Identify outliers for the Middle East
middle_east_score_outliers = identify_outliers(middle_east_data, 'score')
middle_east_gdp_outliers = identify_outliers(middle_east_data, 'Log GDP per capita')

# Print outliers for South Asia
print("South Asia - Score Outliers:")
print(south_asia_score_outliers[['Country name', 'score']])

print("\nSouth Asia - GDP per Capita Outliers:")
print(south_asia_gdp_outliers[['Country name', 'Log GDP per capita']])

# Print outliers for the Middle East
print("\nMiddle East - Score Outliers:")
print(middle_east_score_outliers[['Country name', 'score']])

print("\nMiddle East - GDP per Capita Outliers:")
print(middle_east_gdp_outliers[['Country name', 'Log GDP per capita']])
```

```python
# Function to plot data and highlight outliers
def visualize_data_with_outliers(data, outliers, x, y, title, x_label, y_label):
    plt.figure(figsize=(16, 8))
    
    # Plot regular data
    plt.scatter(data[x], data[y], label='Normal', color='purple')
    
    # Highlight outliers
    plt.scatter(outliers[x], outliers[y], color='red', label='Outliers')
    
    # Add labels and title
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid()
    plt.show()

# Plot South Asia - Score Outliers
visualize_data_with_outliers(
    south_asian_data,
    south_asia_score_outliers,
    x='Country name',
    y='score',
    title='South Asia - Score Outliers',
    x_label='Country',
    y_label='Score'
)

# Plot South Asia - GDP Outliers
visualize_data_with_outliers(
    south_asian_data,
    south_asia_gdp_outliers,
    x='Country name',
    y='Log GDP per capita',
    title='South Asia - GDP Outliers',
    x_label='Country',
    y_label='Log GDP'
)

# Plot Middle East - Score Outliers
visualize_data_with_outliers(
    middle_east_data,
    middle_east_score_outliers,
    x='Country name',
    y='score',
    title='Middle East - Score Outliers',
    x_label='Country',
    y_label='Score'
)

# Plot Middle East - GDP Outliers
visualize_data_with_outliers(
    middle_east_data,
    middle_east_gdp_outliers,
    x='Country name',
    y='Log GDP per capita',
    title='Middle East - GDP Outliers',
    x_label='Country',
    y_label='Log GDP'
)
```
